## 1️⃣ Qu'est-ce que JupyterLab ?

**JupyterLab** est un environnement de développement interactif qui permet de :
- Écrire et exécuter du code Python
- Documenter son travail avec du texte formaté (Markdown)
- Visualiser des données et des graphiques
- Travailler de manière interactive, cellule par cellule

### Pourquoi c'est populaire en Data Science ?
- On peut **exécuter du code pas à pas**
- On peut **voir les résultats immédiatement**
- On peut **mélanger explications et code**
- C'est parfait pour **explorer des données**

## 2️⃣ Les deux types de cellules

### Cellule Markdown (comme celle-ci)
- Contient du **texte formaté**
- Permet d'écrire des titres, listes, liens...
- Ne s'exécute pas comme du code

### Cellule Code (comme celle du dessous)
- Contient du **code Python**
- S'exécute avec `Shift + Enter`
- Affiche le résultat juste en dessous

## 4️⃣ Les raccourcis clavier essentiels

| Raccourci | Action |
|-----------|--------|
| `Shift + Enter` | Exécuter la cellule et passer à la suivante |
| `Ctrl + Enter` | Exécuter la cellule sans bouger |
| `Esc` | Passer en mode commande (bordure bleue) |
| `Enter` | Passer en mode édition (bordure verte) |
| `A` | Insérer une cellule au-dessus (en mode commande) |
| `B` | Insérer une cellule en dessous (en mode commande) |
| `DD` | Supprimer une cellule (en mode commande) |
